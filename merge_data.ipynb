```python
import pandas as pd
import numpy as np
```

```python
# import the data

bacteria_phylum = pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_phylum_scaled.csv")
bacteria_class = pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_class_scaled.csv")
bacteria_order = pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_order_scaled.csv")
bacteria_family = pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_family_scaled.csv")
bacteria_genus = pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_genus_scaled.csv")
bacteria_species= pd.read_csv(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_species_scaled.csv")
bacteria_aerotolerance = pd.read_excel(
    "C:/Users/Jennifer Chen/Documents/course documents/spring 2020/ibd project/data/UNC_WashU_CD_NI_ileum_only_aerotolerance.xlsx")
```

```python
# reset the row names as the sample names

def reset_index(df):
    df.rename(columns = {'Unnamed: 0': 'SampleID'}, inplace=True)
    new_df = df.set_index('SampleID')
    return new_df

bacteria_phylum = reset_index(bacteria_phylum)
bacteria_class = reset_index(bacteria_class)
bacteria_order = reset_index(bacteria_order)
bacteria_family = reset_index(bacteria_family)
bacteria_genus = reset_index(bacteria_genus)
bacteria_species = reset_index(bacteria_species)
bacteria_aerotolerance = reset_index(bacteria_aerotolerance)
```

```python
bacteria_species[:5]
```

```python
# concatenate the dataframes

bacteria_data = pd.concat([bacteria_species.sort_index(), bacteria_genus.sort_index(), 
                           bacteria_family.sort_index(), bacteria_order.sort_index(), 
                           bacteria_class.sort_index(), bacteria_phylum.sort_index(), 
                           bacteria_aerotolerance.sort_index()], axis=1, join='outer')
```
